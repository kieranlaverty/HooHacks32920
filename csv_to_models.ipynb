```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.utils import shuffle

testset = shuffle(testset)

testset
```

```python
import sklearn.linear_model

#prepare the data
y = np.c_[testset['Confirmed']] 
X = np.c_[testset['Past_Confirmed']]

#selecting and training model
model = sklearn.linear_model.LinearRegression()
model.fit(X, y)

x_new = [[814180.0]]
prediction = model.predict(x_new)
print(prediction)


plt.plot(X, y, 'o', color='black');
```

```python
import pickle 

from sklearn.externals import joblib
# Output a pickle file for the model
joblib.dump(model, 'saved_model.pkl')

# Load the pickle file
model_load = joblib.load('saved_model.pkl') 

# Check that the loaded model is the same as the original
assert model.score(X, y) == model_load.score(X, y)


model_load.predict(x_new)
```
